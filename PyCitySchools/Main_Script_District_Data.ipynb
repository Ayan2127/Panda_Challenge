```python
#dependencies
import pandas as pd
from pathlib import Path
```

```python
#idenitify/read csv files, asign variable for each csv file, and display merged csv files in ONE df
school_data_file = "../PyCitySchools/Resources/schools_complete.csv"
student_data_file = "../PyCitySchools/Resources/students_complete.csv"
school_data = pd.read_csv(school_data_file, low_memory =False)
student_data = pd.read_csv(student_data_file, low_memory =False)
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

```python
#Number of different schools using unique function (removies school_name repitition)
list_unique_schools = school_data_complete["school_name"].unique()
num_school_count = school_data_complete["school_name"].nunique()
print(f"Total number of unique schools: {num_school_count}\n\n")
print(f"List of unique schools: \n\n{list_unique_schools}\n\n")
```

```python
#Total students across district found using value counts and sum function
total_student = school_data_complete["student_name"].value_counts().sum()
print(f"Total students enrolled in the district: {total_student}")
```

```python
#Total budget of unique schools summed
total_budget = school_data_complete["budget"].unique().sum()
print(f"The total budget of schools combined: ${total_budget}")
```

```python
#District average math score, combines average math scores of all schools in 'school_name' column listed in the merged csv
average_math_score = school_data_complete["math_score"].mean()
print(f"District average math score: {average_math_score}")
```

```python
#District average reading score, combines average reading scores of all schools in 'school_name' column listed in the merged csv
average_reading_score = school_data_complete["reading_score"].mean()
print(f"District average reading score: {average_reading_score}")
```

```python
#Percentage of students who passed math, in other words achieved greater than 70 points
passing_math_count = school_data_complete[(school_data_complete["math_score"] >= 70)].count()["student_name"]
passing_math_percentage = passing_math_count / float(total_student.sum()) * 100
print(f"Percentage of students who passed math: {passing_math_percentage}%")
```

```python
#Percentage of students who passed reading, in other words achieved greater than 70 points
passing_reading_count = school_data_complete[(school_data_complete["reading_score"] >= 70)].count()["student_name"]
passing_reading_percentage = passing_reading_count / float(total_student.sum()) * 100
print(f"Percentage of students who passed reading: {passing_reading_percentage}%")
```

```python
#Percentage of student who passed BOTH reading and math 
passing_math_reading_count = school_data_complete[
    (school_data_complete["math_score"] >= 70) & (school_data_complete["reading_score"] >= 70)
].count()["student_name"]
overall_passing_rate = passing_math_reading_count /  float(total_student.sum()) * 100
print(f"Percentage of students who passed math & reading: {overall_passing_rate}%")
```

```python
#Summary DataFrame that combines calculations from above cells
district_summary = pd.DataFrame({"Total Schools": [num_school_count], "Total Students":[total_student], "Total Budget" : [total_budget], "Average Math Score": [average_math_score], "Average Reading Score": [average_reading_score], "% Passing Math": [passing_math_percentage], "% Passing Reading": [passing_reading_percentage], "% Overall Passing": [overall_passing_rate]})


district_summary["Total Schools"] = district_summary["Total Schools"].map("{:,}".format)
district_summary["Total Students"] = district_summary["Total Students"].map("{:,}".format)
district_summary["Total Budget"] = district_summary["Total Budget"].map("${:,.2f}".format)
district_summary["Average Math Score"] = district_summary["Average Math Score"].map("{:,.5f}".format)
district_summary["Average Reading Score"] = district_summary["Average Reading Score"].map("{:,.5f}".format)
district_summary["% Passing Math"] = district_summary["% Passing Math"].map("{:,.5f}".format)
district_summary["% Passing Reading"] = district_summary["% Passing Reading"].map("{:,.5f}".format)
district_summary["% Overall Passing"] = district_summary["% Overall Passing"].map("{:,.5f}".format)


district_summary.head()
```

# School Summary

```python
#Type of school per 'school_name'
school_types = school_data.set_index(["school_name"])["type"]
school_types
```

```python
#Total students per school 
##Another solution:total_students_per_school = school_data_complete.groupby("school_name")["size"].count()
total_students_per_school = school_data_complete["school_name"].value_counts()
total_students_per_school
```

```python
#Total budget per school using mean function
budget_total_per_school = school_data_complete.groupby("school_name")["budget"].mean()
budget_total_per_school
```

```python
#Merge calculations of average math and reading scores with index set as school name
per_school_math = school_data_complete.groupby("school_name")["math_score"].mean()
per_school_reading = school_data_complete.groupby("school_name")["reading_score"].mean()
per_school_math_reading = pd.merge(per_school_math, per_school_reading, on="school_name")
per_school_math_reading
```

```python
#Num of students per school with math scores of 70 or higher 
num_students_passing_math_per_school = school_data_complete[(school_data_complete["math_score"] >= 70)]
num_students_passing_math_per_school
```

```python
#Num of students per school with reading scores of 70 or higher 
num_students_passing_reading_per_school = school_data_complete[(school_data_complete["reading_score"] >= 70)]
num_students_passing_reading_per_school
```

```python
#Num of students per school with math and reading scores of 70 or higher 
students_passing_math_and_reading = school_data_complete.loc[(school_data_complete["reading_score"] >= 70) & (school_data_complete["math_score"] >= 70)]
school_students_passing_math_and_reading = students_passing_math_and_reading.groupby(["school_name"]).size()
school_students_passing_math_and_reading
```

```python
#Per student budget 
budget_per_student = budget_total_per_school / total_students_per_school
print(budget_per_student)
```

```python
#Percentages of students passing math, reading & overall
per_school_passing_math = num_students_passing_math_per_school.groupby(["school_name"]).count()["student_name"]/total_students_per_school * 100
per_school_passing_reading = num_students_passing_reading_per_school.groupby(["school_name"]).count()["student_name"]/total_students_per_school * 100
overall_passing_rate = school_students_passing_math_and_reading / total_students_per_school * 100
```

```python
#DataFrame called `per_school_summary` with columns for the calculations above.
per_school_summary = pd.DataFrame({"School Type": school_types, "Total Students": total_students_per_school, "Total School Budget": budget_total_per_school, "Per Student Budget": budget_per_student, "Average Math Score": per_school_math, "Average Reading Score": per_school_reading, "% Passing Math": per_school_passing_math, "% Passing Reading": per_school_passing_reading, "% Overall Passing":overall_passing_rate})
                                   

# Formatting dollar signs for budget
per_school_summary["Total School Budget"] = per_school_summary["Total School Budget"].map("${:,.2f}".format)
per_school_summary["Per Student Budget"] = per_school_summary["Per Student Budget"].map("${:,.2f}".format)

#Finalized DataFrame
per_school_summary
```

# Scores by School Spending
